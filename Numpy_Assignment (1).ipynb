```python
import matplotlib.pyplot as plt
import os
import glob
import matplotlib.image as mpimg
```

```python
import numpy as np
import cv2


image1 = cv2.imread('img1.jpg')
image2 = cv2.imread('img2.jpg')
res1 = cv2.resize(image1, dsize=(120,120))
res2 = cv2.resize(image2, dsize=(120,120))
```

```python
res1=np.ones((120,120,3))+res1
```

```python
res2=np.zeros((120,120,3))+res2
```

```python
ConCat = cv2.vconcat([res1, res2])
cv2.imwrite('Final.jpg', ConCat)
```

```python
Res=ConCat/255
Res
```

```python
plt.imshow(Res)
```
